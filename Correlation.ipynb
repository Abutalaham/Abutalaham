```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create the series
x = pd.Series([1, 2, 3, 4, 5, 6, 7])
y = pd.Series([1, 2, 3, 4, 5, 5, 4])

# Calculate the correlation
correlation = y.corr(x)
print(f"Correlation between x and y: {correlation}")

# Plot the data
plt.plot(x, y)

```
