```python
import pandas as pd
df=pd.read_csv('dev.csv')
df
```

```python
mean = df['experts'].mean()
df['experts'] = df['experts'].fillna(mean)
df
```

```python
mode = df['developer'].mode()
df['developer']=df['developer'].fillna(mode)
df
```

```python
median=df['salary'].median()
df['salary']=df['salary'].fillna(median)
df
```
